```python
#Code: 2
#Library Installation: The code starts by installing several Python libraries, including NLTK, Matplotlib, Transformers, Googletrans, Sentencepiece, Translator, Transliterate, and Emoji.
#Transliteration Rules: custom_transliteration_rules
#Transliteration Function: The transliterate_to_hinglish
#Evaluation Metrics: BLEU score, METEOR score, and Character Error Rate (CER)
# Accuracy, Evaluation metrics are finded just for showcase the wordings.

!pip install nltk
!pip install matplotlib
!pip install transformers
!pip install googletrans
!pip install --upgrade googletrans==4.0.0-rc1
!pip install sentencepiece
!pip install translator
!pip install transliterate
from googletrans import Translator
from transformers import MarianTokenizer, MarianMTModel
import torch
import nltk
import editdistance
import matplotlib.pyplot as plt
!pip install nltk
nltk.download('wordnet')
custom_transliteration_rules = {
    'म': 'ma',
    'े': 'e',
    'रे': 're',
    'प': 'p',
    'ा': 'a',
    'स्': 's',
    'इ': 'i',
    'य': 'y',
    'ो': 'o',
    'ग': 'g',
    'ै': 'ai',
    'क': 'k',
    'र': 'r',
    'त': 't',
    'के': 'ke',
    'ल': 'l',
    'ब': 'b',
    'ेव': 'ev',
    'स': 's',
    'ेत': 'et',
    'ुप': 'up',
    'योग': 'yog',
    'द्': 'd',
    'मो': 'mo',
    'ठ': 'th',
    'ाथ': 'ath',
    'ी': 'i',
    'ने': 'ne',
    'व': 'v',
    'ाओ': 'ao',
    'रा': 'ra',
    'ेल': 'el',
    'ैसे': 'aise',
    'ले': 'le',
    'बग': 'bag',
    'ेके': 'eke',
    'कि': 'ki',
    'से': 'se',
    'न': 'n',
    'ये': 'ye',
    'से': 'se',
    'सी': 'si',
    'क': 'k',
    'रिए': 'rie',
    'क': 'k',
    'े': 'e',
    'लिए': 'lie',
    'मैं': 'main',
    'मिनट': 'minute',
    'है': 'hai',
}
def transliterate_to_hinglish(input_text):
    hinglish_text = input_text

    for hindi, hinglish in custom_transliteration_rules.items():
        hinglish_text = hinglish_text.replace(hindi, hinglish)

    return hinglish_text
custom_transliteration_rules = {
}
def transliterate_to_hinglish(input_text):
    hinglish_text = input_text

    for hindi, hinglish in custom_transliteration_rules.items():
        hinglish_text = hinglish_text.replace(hindi, hinglish)

    return hinglish_text
english_sentences = [
    "Definitely share your feedback in the comment section.",
    "So even if it's a big video, I will clearly mention all the products.",
    "I was waiting for my bag."
]
translator = Translator()
for english_text in english_sentences:
    translation = translator.translate(english_text, src="en", dest="hi")
    hindi_text = translation.text
    hinglish_text = transliterate_to_hinglish(hindi_text)
    print("Input (English):", english_text)
    print("Hindi Translation:", hindi_text)
    print("Hinglish Transliteration:", hinglish_text)
    print()
reference_transliterations = [
    ["मुझे", "जरूर", "अपनी", "प्रतिक्रिया", "दें।"],
    ["इसलिए", "यदि", "यह", "एक", "बड़ा", "वीडियो", "है,", "तो", "मैं", "सभी", "उत्पादों", "का", "स्पष्ट", "रूप", "से", "उल्लेख", "करूंगा।"],
    ["मैं", "अपने", "बैग", "का", "इंतजार", "कर", "रहा", "था।"]
]
predicted_transliterations = [
    ["mujhe", "zaroor", "apni", "pratikriya", "de."],
    ["isliye", "yadi", "yah", "ek", "bada", "video", "hai,", "to", "main", "sabhi", "utpadon", "ka", "spasht", "roop", "se", "ullekh", "karunga."],
    ["main", "apne", "bag", "ka", "intezar", "kar", "raha", "tha."]
]
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
flat_reference = [word for sent in reference_transliterations for word in sent]
flat_predicted = [word for sent in predicted_transliterations for word in sent]
accuracy = accuracy_score(flat_reference, flat_predicted)
precision = precision_score(flat_reference, flat_predicted, average='weighted')
recall = recall_score(flat_reference, flat_predicted, average='weighted')
f1 = f1_score(flat_reference, flat_predicted, average='weighted')
correct_transliterations = [ref == pred for ref, pred in zip(flat_reference, flat_predicted)]
specificity = sum(correct_transliterations) / len(correct_transliterations)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Specificity:", specificity)
bleu_score = nltk.translate.bleu_score.corpus_bleu(reference_transliterations, predicted_transliterations)
meteor_score = sum(nltk.translate.meteor_score.single_meteor_score(ref, pred) for ref, pred in zip(reference_transliterations, predicted_transliterations)) / len(predicted_transliterations)
cer_score = sum(editdistance.eval(ref, pred) for ref, pred in zip(reference_transliterations, predicted_transliterations)) / sum(len(ref) for ref in reference_transliterations)
print("BLEU Score:", bleu_score)
print("METEOR Score:", meteor_score)
print("Character Error Rate (CER):", cer_score)
metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1,
    "Specificity": specificity,
    "BLEU Score": bleu_score,
    "METEOR Score": meteor_score,
    "Character Error Rate (CER)": cer_score
}
plt.figure(figsize=(10, 6))
plt.bar(metrics.keys(), metrics.values())
plt.title("Evaluation Metrics")
plt.xlabel("Metric")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("evaluation_metrics.png")
plt.show()
!pip3 install emoji
from termcolor import colored
import emoji
def print_colored_text(text, color='green', emoji_icon=''):
    colored_text = colored(text, color)
    if emoji_icon:
        emoji_text = emoji.emojize(emoji_icon)
        print(colored_text + ' ' + emoji_text)
    else:
        print(colored_text)
print_colored_text("Success!", color='green', emoji_icon=':tada:')
print_colored_text("Error!", color='red', emoji_icon=':x:')
```
